```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import SGD
import matplotlib.pyplot as plt
```

### Split and Scale the Fasion MNIST Data into Train and Test Sets

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()


input_dim = 28 * 28 # pixel length and width of each image
output_dim = nb_classes = 10
nb_epoch = 20

x_train = x_train.reshape(60000, input_dim)
x_test = x_test.reshape(10000, input_dim)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255 # normalize by dividing by maximum pixel value
x_test /= 255
```

```python
y_train = to_categorical(y_train, nb_classes)
y_test = to_categorical(y_test, nb_classes)
```

### Define the Parameters to be Tuned for Each Deep Learning Model

```python
def fit_model(x_train=x_train, x_test=x_test, y_train=y_train, y_test=y_test,
              activation='relu', optimizer='sgd', loss='categorical_crossentropy', batch_size=128):
    
    model = Sequential()
    model.add(Dense(1028, input_shape=x_train[0].shape, activation=activation))
    model.add(Dense(1024, activation=activation))
    model.add(Dense(512, activation=activation))
    model.add(Dense(256, activation=activation))
    model.add(Dense(128, activation=activation))
    model.add(Dense(64, activation=activation))
    model.add(Dense(10, activation='softmax'))
    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    history = model.fit(x_train, y_train, batch_size=batch_size, epochs=15, verbose=1, validation_data=(x_test, y_test))
    parameters = {'activation':activation, 'optimizer':optimizer, 'loss':loss, 'batch_size':batch_size}
    return history, parameters
```

```python
activations = ['relu', 'tanh', 'sigmoid']
optimizers = ['sgd', 'adam']
losses = ['categorical_crossentropy', 'hinge']
batches = [64, 128, 256]
```

### Fit the Models and Organize the Data

```python
history_list = []
parameters_list = []
for a in activations:
    for o in optimizers:
        for l in losses:
            for b in batches:
                history, parameters = fit_model(activation=a, optimizer=o, loss=l, batch_size=b)
                history_list.append(history)
                parameters_list.append(parameters)
```

```python
d_list1 = []
d_list2 = []
for i in range(len(parameters_list)):
    dict1 = {'accuracy':history_list[i].history['accuracy'][-1],
                'val_accuracy':history_list[i].history['val_accuracy'][-1],
                'loss':history_list[i].history['loss'][-1],
                'val_loss':history_list[i].history['val_loss'][-1]}
    d_list1.append(dict1)
    d_list2.append(parameters_list[i])
    
df1 =  pd.DataFrame(d_list1)
df2 = pd.DataFrame(d_list2)

df3 = pd.concat([df1, df2], axis=1)
```

```python
df3['diff_accuracies_%'] = 100 * abs(df3['accuracy'] - df3['val_accuracy']) / df3['val_accuracy']
df3.sort_values('val_accuracy', ascending=False, inplace=True)
df3.head(10)
```

From the above results, it appears the specific combination of: activation=relu, optimizer=adam, loss=categorical_crossentropy was the strongest parameter tunings for the first 3 models with a batch size of 128 being the best tuning out of those models. If you wanted to sacrifice accuracy a hair for slightly better confidence that your models werent overfit, the model with activation=tanh and batch_size=256 was the next best performer with ~88.6% accuracy. For comparison, the best model only had an accuracy of ~89.2%, so the differences between model performance is pretty negligible as they are both fairly accurate.
